```python
import random
```

```python
def run_simulation(level, attempts):
  maxlv = 0
  while attempts != 0:
    if level == 2:
      level = level + random.choices(([1, 0]),weights=[70,30], k=1)[0]

    elif level == 3:
      level = level + random.choices(([1, -1]),weights=[55,45], k=1)[0]

    elif level == 4:
      level = level + random.choices(([1, -1]),weights=[40,60], k=1)[0]

    elif level == 5:
      level = level + random.choices(([1, -1]),weights=[30.7,69.30], k=1)[0]

    elif level == 6:
      level = level + random.choices(([1, -1, -4]),weights=[20.5, 76.5, 3], k=1)[0]

    elif level == 7:
      level = level + random.choices(([1, -1, -5]),weights=[15.3, 81.7, 3], k=1)[0]

    elif level == 8:
      level = level + random.choices(([1, -1, -6]),weights=[10,87,3], k=1)[0]

    elif level == 9:
      maxlv = max(maxlv, level)
      print("Reached level 9 after ", (100-attempts), "attempts")
      break
    attempts = attempts - 1
    maxlv = max(maxlv, level)
  print("Max level reached was ", maxlv)
```

```python

run_simulation(2, 100)
```

```python
def run_simulation2(level, attempts):
  maxlv = 0
  while attempts != 0:
    if level == 2:
      level = level + random.choices(([1, 0]),weights=[70,30], k=1)[0]

    elif level == 3:
      level = level + random.choices(([1, -1]),weights=[55,45], k=1)[0]

    elif level == 4:
      level = level + random.choices(([1, -1]),weights=[40,60], k=1)[0]

    elif level == 5:
      level = level + random.choices(([1, -1]),weights=[30.7,69.30], k=1)[0]

    elif level == 6:
      level = level + random.choices(([1, -1, -4]),weights=[20.5, 76.5, 3], k=1)[0]

    elif level == 7:
      level = level + random.choices(([1, -1, -5]),weights=[15.3, 81.7, 3], k=1)[0]

    elif level == 8:
      level = level + random.choices(([1, -1, -6]),weights=[10,87,3], k=1)[0]

    elif level == 9:
      maxlv = max(maxlv, level)
      break
    attempts = attempts - 1
    maxlv = max(maxlv, level)
  return maxlv
```

```python
average = 0
tries = 10000
while tries != 0:
  average = average + run_simulation2(2, 30)
  tries = tries-1
final_average = average/10000

print(final_average)
```
